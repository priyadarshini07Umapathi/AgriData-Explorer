```python
current_year = df['Year'].max()
# Calculate the start year for the last 50 years
start_year = current_year - 49
df_last_50_years = df[df['Year'] >= start_year]
df_last_50_years
```

```python
df_melted = pd.melt(df_last_50_years, id_vars=['Year'], value_vars=['RICE_PRODUCTION_1000_tons', 'WHEAT_PRODUCTION_1000_tons'], var_name='Metric', value_name='Amount')

# Create the grouped bar chart using seaborn.barplot()
pt.figure(figsize=(8, 6))
Year=range(2017,1968)
sns.barplot(x='Year', y='Amount', hue='Metric', data=df_melted, palette='viridis')

pt.title('Rice Production vs Wheat Production')
pt.xlabel('Year')
pt.ylabel('Production in Tons')
pt.xticks(rotation=45, ha='right')
pt.legend(title='Metric')
pt.show()
```

```python
selected_columns=['Year','RICE_AREA_1000_ha','RICE_PRODUCTION_1000_tons','WHEAT_AREA_1000_ha','WHEAT_PRODUCTION_1000_tons','MAIZE_AREA_1000_ha','MAIZE_PRODUCTION_1000_tons']
filtered_df_RWM=df[selected_columns]
filtered_df_RWM['Decade'] = (filtered_df_RWM['Year'] // 10) * 10
filtered_df_RWM=filtered_df_RWM.drop('Year', axis=1)
filtered_df_RWM
```

```python
df_RWM = pd.melt(filtered_df_RWM,id_vars=['Decade'],var_name='Metric',value_name='Value')
pt.figure(figsize=(12, 6))
g = sns.catplot(
    data=df_RWM, 
    kind="bar",
    x="Decade", 
    y="Value", 
    hue="Metric",
    palette="dark",
    height=6, 
    aspect=1.5
)
# Customizing the plot
g.despine(left=True)
g.set_axis_labels("Decade", "Value")
g.legend.set_title("Metrics")
pt.title("Grouped Bar Chart of 6 Metrics by Year")
pt.xticks(rotation=45) # Rotate x-axis labels for better readability if needed
pt.show()
```
